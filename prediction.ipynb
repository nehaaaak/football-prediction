```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

### Feature Correlation Matrix

```python
# Correlation matrix of predictor features
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(train[predictors].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Predictor Features")
plt.show()
```

This heatmap shows the correlation between selected features like venue, opponent, match hour, and match day. It helps us identify if any features are highly correlated and potentially redundant.

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
error = accuracy_score(test["target"], preds)
```

```python
from sklearn.metrics import precision_score

precision_score(test["target"], preds)
```

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
error
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
combined.head(10)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```

```python
merged
```

```python
merged[(merged["predicted_x"] == 1) & (merged["predicted_y"] ==0)]["actual_x"].value_counts()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# âœ… Visualize correct vs incorrect predictions (Team X)
merged["Correct"] = merged["actual_x"] == merged["predicted_x"]

sns.countplot(x="Correct", data=merged, palette="pastel")
plt.title("Correct vs Incorrect Predictions (Team X)")
plt.xlabel("Prediction Correct?")
plt.ylabel("Number of Matches")
plt.xticks([0, 1], ["Incorrect", "Correct"])
plt.show()
```

This bar chart visualizes how many of the model's predictions were correct versus incorrect for Team X.

### Accuracies

```python
# For team_x side
print("âœ… Accuracy for Team X:", accuracy_score(merged["actual_x"], merged["predicted_x"]))
print("\nðŸ“‹ Classification Report for Team X:\n")
print(classification_report(merged["actual_x"], merged["predicted_x"]))
```

The accuracy and classification report give insight into how well the model predicts match outcomes.
